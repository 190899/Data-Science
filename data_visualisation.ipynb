```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("DataScience_Salary.csv")
dataset
```

```python
#print no of rows and columns
print("No of rows:",dataset.shape[0])
print("No of columns:",dataset.shape[1])
```

```python
import seaborn as sb
```

```python
sb.distplot(dataset['work_year'],hist=True,kde=True, kde_kws={'color':'red'})
plt.show()
```

```python
sb.jointplot(x='work_year',y='salary',data=dataset)
plt.show()
```

```python
sb.jointplot(x='work_year',y='salary',data=dataset,kind="hex")
plt.show()
```

```python
#pair plot
sb.pairplot(dataset,hue='employment_type',diag_kind="kde",kind="scatter",palette="husl")
plt.show()
```

```python
#categorical plot
sb.stripplot(x='employment_type',y='salary_in_usd',data=dataset)
plt.show()
```

```python
sb.swarmplot(x='employment_type',y='salary_in_usd',data=dataset)
plt.show()
```

```python
sb.violinplot(x='employment_type',y='salary_in_usd',data=dataset)
plt.show()
```

```python
sb.boxplot(x='employment_type',y='salary_in_usd',data=dataset,hue='experience_level')
plt.show()
```

```python
sb.barplot(x='employment_type',y='salary_in_usd',data=dataset,hue='experience_level')
plt.show()
```

```python
sb.countplot(x='employment_type',data=dataset,hue='experience_level')
plt.show()
```

```python
sb.pointplot(x='salary_in_usd',y='employment_type',data=dataset,hue='experience_level')
plt.show()
```

```python
sb.boxplot(data=dataset,orient='h')
plt.show()
```
