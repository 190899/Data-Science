```python
import pandas as pd
```

```python
dataset=pd.read_csv("Data_Science_Salary.csv")
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset.fillna(0)
```

```python
quan=[]
qual=[]
for columnsName in dataset:
    if dataset[columnsName].dtype=='O':
        qual.append(columnsName)
    else:
        quan.append(columnsName)
```

```python
quan
```

```python
qual
```

```python
dataset[quan]
```

```python
dataset[quan].fillna(0)
```

```python
dataset[quan].isnull().sum()
```

```python
dataset[qual].isnull().sum()
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(dataset[quan])
df=imp.transform(dataset[quan])
df=pd.DataFrame(df,columns=quan)
```

```python
df
```

```python
df[quan].isnull().sum()
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan, strategy='constant')
imp.fit(dataset[qual])
df=imp.transform(dataset[qual])
df=pd.DataFrame(df,columns=qual)
```

```python
df
```

```python
cate=df[qual]
```

```python
combine=[df,cate]
```

```python
df[qual].isnull().sum()
```

```python
preprocessing=pd.concat(combine,axis=1)
```

```python
preprocessing.to_csv("PreDataScienceSalary.csv")
```
